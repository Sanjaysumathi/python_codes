DECISION TREE WITH IRIS DATASET

```
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```
iris = pd.read_csv("iris.csv")
```

```

iris.head()
```

```
iris.info()
```

SPLITTING DATASETS AS X AND Y


```
X=iris.iloc[:,:-1].values
y=iris.iloc[:,-1].values
```

```
x_train,x_test,y_train,y_test= train_test_split(X,y,test_size=0.2, random_state=1)
```

MODELLING THE DECISION TREE

```
reg=DecisionTreeClassifier()
mod= reg.fit(x_train,y_train)
```

```
y_pred= mod.predict(x_test)
```

```
mod.score(x_test,y_test)
```

MODEL EVALVATION

```
result=pd.DataFrame({"y_pred":y_pred, "y_test":y_test})
result.head()
```

VISUALIZATION OF MODEL

```
fig= plt.figure(figsize=(15,15))
_ = plot_tree(reg,feature_names= iris.columns[:-1], class_names= np.unique(y), filled=True)
```
